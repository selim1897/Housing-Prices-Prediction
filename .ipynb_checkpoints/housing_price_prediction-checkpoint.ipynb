# Libraries

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.impute import SimpleImputer

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

# Preprocessing and Data Analysis

```python
train = pd.read_csv("train.csv")
train.head()
```

```python
train.info()
```

```python
# Dropping columns with more than 50% missing values and the Id column
cols_to_drop = ['Id', 'Alley', 'MasVnrType', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature']
train.drop(cols_to_drop, axis=1, inplace=True)
```

```python
# get the numerical columns
numeric = train.select_dtypes(exclude=['object']).drop(['MSSubClass'], axis=1).copy()
numeric.columns
```

```python
# get the categorical columns
cat_train = train.select_dtypes(include=['object']).copy()
cat_train['MSSubClass'] = train['MSSubClass']   #MSSubClass is nominal
cat_train.columns
```

## Univariate Analysis

### Numerical Columns visualizations

```python
fig = plt.figure(figsize=(18,16))
for index,col in enumerate(numeric.columns):
    plt.subplot(9,4,index+1)
    sns.distplot(numeric.loc[:,col].dropna(), kde=False)
fig.tight_layout(pad=1.0)
```

```python
# Removing features that have mostly just 1 value
numeric.drop(['LowQualFinSF', '3SsnPorch', 'PoolArea', 'MiscVal', 'ScreenPorch', 'EnclosedPorch', 'KitchenAbvGr', 
            'BsmtHalfBath', 'BsmtFinSF2'], axis=1, inplace=True)
```

```python
fig = plt.figure(figsize=(14,15))
for index,col in enumerate(numeric.columns):
    plt.subplot(7,4,index+1)
    sns.boxplot(y=col, data=numeric.dropna())
fig.tight_layout(pad=1.0)
```

### Categorical columns visualizations

```python
fig = plt.figure(figsize=(18,20))
for index in range(len(cat_train.columns)):
    plt.subplot(8,5,index+1)
    sns.countplot(x=cat_train.iloc[:,index], data=cat_train.dropna())
    plt.xticks(rotation=90)
fig.tight_layout(pad=1.0)
```

```python
# Removing features that have mostly just 1 value
removed_cols = []
for col in cat_train.columns:
    if col != "MSSubClass":
        if cat_train[col].value_counts()[0] > 1150:
            removed_cols.append(col)
            cat_train.drop(col, axis=1, inplace=True)

removed_cols
```

## Bi-Variate Analysis

```python
plt.figure(figsize=(14,12))
correlation = numeric.corr()
sns.heatmap(correlation, mask = correlation <0.5, linewidth=0.5, cmap='Blues')
```

```python
# removing the highly correlated features to avoid the problem of multicollinearity
numeric.drop(['GarageYrBlt','TotRmsAbvGrd','1stFlrSF','GarageCars'], axis=1, inplace=True)
```

```python
numeric.corr()["SalePrice"].sort_values(ascending=False)
```

```python
fig = plt.figure(figsize=(20,20))
for index in range(len(numeric.columns)):
    plt.subplot(5,5,index+1)
    sns.scatterplot(x=numeric.iloc[:,index], y='SalePrice', data=numeric.dropna())
fig.tight_layout(pad=1.0)
```

```python
numeric.drop(['MoSold','YrSold'], axis=1, inplace=True)
```

```python
train = train[list(numeric.columns) + list(cat_train.columns)]
```

```python
train.shape
```

# Pipelines

```python
ordinal_cols = ['OverallQual', 'OverallCond', 'ExterQual', 'BsmtQual', 'BsmtExposure', 'BsmtFinType1', 'HeatingQC', 'KitchenQual']
cat_cols = [col for col in cat_train.columns if col not in ordinal_cols]
cat_cols
```

```python
ordinal_map = ['NA', 'Po', 'Fa', 'TA', 'Gd', 'Ex']
fintype_map = ['NA', 'Unf', 'LwQ', 'Rec', 'BLQ', 'ALQ', 'GLQ']
expose_map = ['NA', 'No', 'Mn', 'Av', 'Gd']
```

```python
num_pipe = Pipeline([('imputer', SimpleImputer(missing_values=pd.NA, strategy='mean')), 
                 ('scaler', MinMaxScaler())])

cat_pipe = Pipeline([('imputer', SimpleImputer(missing_values=pd.NA, strategy='most_frequent')), 
                 ('OneHotEncoder', OneHotEncoder(handle_unknown='ignore'))])

ordinal_cat_pipe = Pipeline([('imputer', SimpleImputer(missing_values=pd.NA, strategy='most_frequent')),
                            ('ord_enc', OrdinalEncoder(categories=[ordinal_map, ordinal_map, fintype_map, ordinal_map, ordinal_map, expose_map]))])

preprocessor = ColumnTransformer([
    ('numerical_cols', num_pipe, numeric.drop("SalePrice", axis=1).columns),
    ('categorical_cols', cat_pipe, cat_cols),
    ('ordinal_cols', ordinal_cat_pipe, list(set(ordinal_cols) - set(('OverallQual', 'OverallCond')))),
    #('ordinal_enc', OrdinalEncoder(categories=[ordinal_map, ordinal_map, fintype_map, ordinal_map, ordinal_map, expose_map]), list(set(ordinal_cols) - set(('OverallQual', 'OverallCond'))))
])
```

```python
X = train.drop("SalePrice", axis=1).copy()
y = train["SalePrice"].copy()
```

```python
X_prepared = preprocessor.fit_transform(X)
```

# Model Training

```python
ln_reg = LinearRegression()

ln_scores = cross_val_score(ln_reg, X_prepared, y, cv=6, scoring="neg_mean_squared_error")

print("scores:", np.sqrt(-ln_scores))
print("Mean:", np.sqrt(-ln_scores).mean())
print("Standard deviation", np.sqrt(-ln_scores).std())
```

```python
rf_reg = RandomForestRegressor()

rf_scores = cross_val_score(rf_reg, X_prepared, y, cv=6, scoring="neg_mean_squared_error")

print("scores:", np.sqrt(-rf_scores))
print("Mean:", np.sqrt(-rf_scores).mean())
print("Standard deviation", np.sqrt(-rf_scores).std())
```

# Fine-Tune the Model

```python
 param_grid = [
 {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
 {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
 ]

 forest_reg = RandomForestRegressor()
 grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error', return_train_score=True)
 grid_search.fit(X_prepared, y)
```

```python
grid_search.best_params_
```

```python
final_model = grid_search.best_estimator_
```

# Prediction

```python
test = pd.read_csv("test.csv")
test.head()
```

```python
X_test_prepared = preprocessor.transform(test[X.columns])
```

```python
y_pred = final_model.predict(X_test_prepared)
y_pred
```

```python
result = pd.DataFrame({"Id":test["Id"], "SalePrice":y_pred})
result
```

```python
result.to_csv('result.csv', index=False) 
```
